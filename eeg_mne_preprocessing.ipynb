```python
import mne
import os
mne.utils.set_config('MNE_USE_CUDA', 'true')
import matplotlib.pyplot as plt
# %matplotlib qt
# %matplotlib inline
import copy
import asrpy
from tqdm import tqdm
import autoreject
```

```python
# 각 파일명 불러오기
eeg_filelist = os.listdir('../../data/eeg_data/')
eeg_filelist.sort()
print(eeg_filelist)
```

```python
# 파일명 중에 공백 2개인 것 바꿈
eeg_filelist_modify = [name.replace('  ', ' ') for name in eeg_filelist]
```

```python
# Raw 객체로 읽어오기
raws = [mne.io.read_raw_edf('../../data/eeg_data/' + f, exclude=['EEG 23A-23R', 'EEG 24A-24R'], preload=True, verbose=0) for f in eeg_filelist]
```

* Montage 설정을 위한 EEG 채널명 변경

```python
# 채널명 매핑
channel_names = {'EEG Fp1-LE':'Fp1',
                 'EEG F3-LE':'F3',
                 'EEG C3-LE':'C3',
                 'EEG P3-LE':'P3',
                 'EEG O1-LE':'O1',
                 'EEG F7-LE':'F7',
                 'EEG T3-LE':'T3',
                 'EEG T5-LE':'T5',
                 'EEG Fz-LE':'Fz',
                 'EEG Fp2-LE':'Fp2',
                 'EEG F4-LE':'F4',
                 'EEG C4-LE':'C4',
                 'EEG P4-LE':'P4',
                 'EEG O2-LE':'O2',
                 'EEG F8-LE':'F8',
                 'EEG T4-LE':'T4',
                 'EEG T6-LE':'T6',
                 'EEG Cz-LE':'Cz',
                 'EEG Pz-LE':'Pz',
                 'EEG A2-A1':'A2'}
```

## 2. 필터링

- ASR 진행 전 하이패스 필터

```python
# # 50Hz 노치 필터
filtered_eeg = {name:raw.copy().notch_filter(freqs=50, verbose=0) for name, raw in raw_copy.items()}

# high-pass filter
filtered_eeg = {name:raw.copy().filter(0.5, 70, fir_design='firwin', verbose=0) for name, raw in filtered_eeg.items()}
```

## 3. ASR

```python
asr_eeg = {}
for name, raw in tqdm(filtered_eeg.items()):
    asr = asrpy.ASR(sfreq=raw.info['sfreq'], cutoff=25)
    asr.fit(raw)
    asr_eeg[name] = asr.transform(raw)
```

```python
os.makedirs('../../data/asr+bandpass', exist_ok=True)
for name, epoch in asr_eeg.items():
    epoch.to_data_frame().to_csv(f'../../data/asr+bandpass/{name}.csv')
```

## 4. Epoching & Autoreject

```python
# ASR 없이 Autoreject 만
# 이벤트 생성 및 에포크 생성
epochs_dict = {}
for name, sig in filtered_eeg.items():
    # 사용할 채널 선택(eeg만 사용)
    picks = mne.pick_types(sig.info, eeg=True)
    events = mne.make_fixed_length_events(sig, start=0, duration=4, overlap=0) # 4초 마다 이벤트 생성
    tmin = 0.0 # 이벤트 기준 0초
    tmax = 4.0 # 이벤트 기준 +4초
    epochs = mne.Epochs(sig, events=events, tmin=tmin, tmax=tmax, baseline=None,
                        preload=True, picks=picks, reject=None) # 총 4초 길이의 에포크로 나눔
    epochs_dict[name] = epochs
```

## 1. 파일 불러오기

```python
# 각 파일명 불러오기
eeg_filelist = os.listdir('../../data/eeg_data/')
eeg_filelist.sort()
print(eeg_filelist)
```

```python
# TASK 제외
eeg_filelist = [file for file in eeg_filelist if 'TASK' not in file]
```

```python
# 파일명 중에 공백 2개인 것 바꿈
eeg_filelist_modify = [name.replace('  ', ' ') for name in eeg_filelist]
```

```python
# Raw 객체로 읽어오기
raws = [mne.io.read_raw_edf('../../data/eeg_data/' + f, exclude=['EEG 23A-23R', 'EEG 24A-24R'], preload=True, verbose=0) for f in eeg_filelist]
```

```python
raws_dict = {eeg_filelist_modify[i]:raws[i] for i in range(len(raws))}
```

* Montage 설정을 위한 EEG 채널명 변경

```python
# 채널명 매핑
channel_names = {'EEG Fp1-LE':'Fp1',
                 'EEG F3-LE':'F3',
                 'EEG C3-LE':'C3',
                 'EEG P3-LE':'P3',
                 'EEG O1-LE':'O1',
                 'EEG F7-LE':'F7',
                 'EEG T3-LE':'T3',
                 'EEG T5-LE':'T5',
                 'EEG Fz-LE':'Fz',
                 'EEG Fp2-LE':'Fp2',
                 'EEG F4-LE':'F4',
                 'EEG C4-LE':'C4',
                 'EEG P4-LE':'P4',
                 'EEG O2-LE':'O2',
                 'EEG F8-LE':'F8',
                 'EEG T4-LE':'T4',
                 'EEG T6-LE':'T6',
                 'EEG Cz-LE':'Cz',
                 'EEG Pz-LE':'Pz',
                 'EEG A2-A1':'A2'}
```

## 2. 필터링

```python
# 레퍼런스 초기화
raw_copy = {name:raw.copy().set_eeg_reference([], verbose=0) for name, raw in raw_copy.items()}
```

- FASTER 진행 전 하이패스 필터

```python
# high-pass filter
filtered_eeg = {name:raw.copy().filter(1, None, fir_design='firwin', verbose=0) for name, raw in raw_copy.items()}

# 50Hz 노치 필터
filtered_eeg = {name:raw.copy().notch_filter(freqs=50, verbose=0) for name, raw in filtered_eeg.items()}
```

```python
# 이벤트 생성 및 에포크 생성
epochs_dict = {}
for name, raw in filtered_eeg.items():
    events = mne.make_fixed_length_events(raw, start=0, duration=4, overlap=0) # 4초 마다 이벤트 생성
    tmin = 0.0
    tmax = 4.0
    epochs = mne.Epochs(raw, events=events, tmin=tmin, tmax=tmax, baseline=None,
                        preload=True) # 총 4초 길이의 에포크로 나눔
    epochs_dict[name] = epochs
```

```python
epochs_faster = {}
for name, epochs in tqdm(epochs_dict.items()):
    # 1. 불량 채널 확인 및 보간
    epochs.info['bads'] = find_bad_channels(epochs, eeg_ref_corr=False) # eeg_ref_corr은 전극 레퍼런스 있으면 True, average면 False
    if len(epochs.info['bads']) > 0:
        epochs.interpolate_bads()

    # 2. 불량 epoch 확인 및 제거
    bad_epochs = find_bad_epochs(epochs)
    if len(bad_epochs) > 0:
        epochs.drop(bad_epochs)
    
    # 3. 불량 ICA components 확인 및 제거
    ica = mne.preprocessing.ICA(0.99, method='infomax').fit(epochs)
    ica.exclude = find_bad_components(ica, epochs,
                                      use_metrics=['kurtosis', 'power_gradient', 'hurst', 'median_gradient']) # EOG 데이터가 있으면 'eog_correlation'까지 사용 가능
    ica.apply(epochs)
    # ICA 변환 이루 re-baseline 필요
    epochs.apply_baseline(epochs.baseline)

    # 4. 각 epoch 별 불량 채널 확인 및 보간
    bad_channels_per_epoch = find_bad_channels_in_epochs(epochs, eeg_ref_corr=False)
    for i, b in enumerate(bad_channels_per_epoch):
        if len(b) > 0:
            ep = epochs[i]
            ep.info['bads'] = b
            ep.interpolate_bads()
            epochs._data[i, :, :] = ep._data[0, :, :]
    
    # average 레퍼런스로 변환
    epochs.set_eeg_reference('average')
    epochs_faster[name] = epochs
```

```python
# for k in epochs_faster.keys():
#     print(f'{k:10s}  diff:, {len(epochs_dict[k]) - len(epochs_faster[k]):3d}')
```

## 3. 저장

```python
os.makedirs('../../data/faster', exist_ok=True)
for name, epoch in epochs_faster.items():
    epoch.to_data_frame().to_csv(f'../../data/faster/{name}.csv')
```
